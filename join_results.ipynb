```python
%reload_ext autotime
import pandas as pd
pd.set_option('display.max_columns', None)
#pd.set_option("display.max_colwidth", 200)
import json
from glob import glob
from tqdm.auto import tqdm
import os
files = glob("results/*.json")
len(files)
```

```python
results = []
for filename in tqdm(files):
    with open(filename) as f:
        try:
            data = json.load(f)
        except:
            print(f"Can't read {filename}")
            continue
            #os.remove(filename)
    data["AI_description"] = data["description"]
    metadata = filename.replace("results/", "videos/").replace(".result.json", ".info.json")
    with open(metadata) as f:
        meta = json.load(f)
    keys = ['id', 'extractor', 'channel', 'channel_id', 'uploader', 'uploader_id', 'title', 'description', 'timestamp', 'view_count', 'like_count', 'comment_count', 'duration', 'upload_date', 'webpage_url']
    for key in keys:
        data[key] = meta.get(key, None)
    results.append(data)
df = pd.DataFrame(results).sort_values("upload_date")
df
```
