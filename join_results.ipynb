```python
%reload_ext autotime
import pandas as pd
pd.set_option('display.max_columns', None)
#pd.set_option("display.max_colwidth", 200)
import json
from glob import glob
from tqdm.auto import tqdm
import os
from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired
import plotly.io as pio
import plotly.express as px # Plotting
pio.renderers.default = 'notebook'
pd.options.plotting.backend = "plotly"
files = glob("results/*.json")
len(files)
```

```python
results = []
for filename in tqdm(files):
    with open(filename) as f:
        try:
            data = json.load(f)
        except:
            print(f"Can't read {filename}")
            continue
            #os.remove(filename)
    data["AI_description"] = data["description"]
    metadata = filename.replace("results/", "videos/").replace(".result.json", ".info.json")
    with open(metadata) as f:
        meta = json.load(f)
    keys = ['id', 'extractor', 'channel', 'channel_id', 'uploader', 'uploader_id', 'title', 'description', 'timestamp', 'view_count', 'like_count', 'comment_count', 'duration', 'upload_date', 'webpage_url']
    for key in keys:
        data[key] = meta.get(key, None)
    results.append(data)
df = pd.DataFrame(results).sort_values("upload_date")
df
```

```python
df = df[df.timeout]
df
```

```python
df["text"] = df.title.str.cat(df[["description", "transcript"]].astype(str), sep=" ")
df
```

```python
topic_model = BERTopic(representation_model=KeyBERTInspired(), min_topic_size=3, verbose=True)
topic_model.fit(df.text)
topic_model.get_topic_info()
```

```python
topic_model.visualize_topics()
```

```python
df.index = pd.to_datetime(df.timestamp, unit='s')
df = df.loc[df.index.dropna()]
df
```

```python
topics_over_time = topic_model.topics_over_time(df.text, df.index, nr_bins=10)
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=20, height=600)
```
