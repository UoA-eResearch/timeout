```python
results = []for filename in tqdm(files):    with open(filename) as f:        try:            AI_data = json.load(f)        except:            print(f"Can't read {filename}")            continue            #os.remove(filename)    AI_data["AI_description"] = AI_data.pop("description")    metadata = filename.replace("results/", "videos/").replace(".result.json", ".info.json")    with open(metadata) as f:        meta = json.load(f)    keys = ['id', 'extractor', 'channel', 'channel_id', 'uploader', 'uploader_id', 'title', 'description', 'timestamp', 'view_count', 'like_count', 'comment_count', 'duration', 'upload_date', 'webpage_url']    result = {}    for key in keys:        result[key] = meta.get(key, None)    result.update(AI_data)    results.append(result)df = pd.DataFrame(results).sort_values("upload_date")df
```

```python
df.menopause = df.menopause.astype(str)
```

```python
df = df[df.menopause == "True"]df
```

```python
df["text"] = df.title.str.cat(df[["description", "transcript"]].astype(str), sep=" ")df
```

```python
df.index = pd.to_datetime(df.timestamp, unit='s')df = df.loc[df.index.dropna()]df
```

```python
topic_model = BERTopic(representation_model=KeyBERTInspired(), min_topic_size=3, verbose=True)topic_model.fit(df.text.str.replace(r"(menopausal|menopause|supplement|vitamin)", "", regex=True, case=False))topic_model.get_topic_info()
```

```python
topic_model.visualize_topics()
```

```python
topics_over_time = topic_model.topics_over_time(df.text, df.index, nr_bins=10)topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=20, height=600)
```
